```python
df = pd.read_csv(r"C:\Users\zaiid\OneDrive\Desktop\ironhack_final\netflix\netflix_titles.csv")
```

```python
netflix_titles = pd.concat([df, df2])
```

```python
netflix_titles.isna().sum()
```

```python
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Kids' TV"), 'listed_in'] = "Kids' TV"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Children & Family Movies|Music & Musicals"), 'listed_in'] = "Children & Family Movies"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Stand-Up Comedy"), 'listed_in'] = "Stand-Up Comedy"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Comedies, Horror Movies"), 'listed_in'] = "Comedies, Horror Movies"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("LGBTQ"), 'listed_in'] = "LGBTQ Movies"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Comedies") & netflix_titles['listed_in'].str.contains("Dramas"), 'listed_in'] = "Comedies, Dramas"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Dramas") & ~netflix_titles['listed_in'].str.contains("Comedies"), 'listed_in'] = "Dramas"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Comedies") & ~netflix_titles['listed_in'].str.contains("Dramas"), 'listed_in'] = "Comedies"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Horror Movies|TV Horror|Horror"), 'listed_in'] = "Horror"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Romantic Movies|Romantic TV Shows|Anime Features"), 'listed_in'] = "Romantic Movies"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Action & Adventure|Action|Sci-Fi & Fantasy"), 'listed_in'] = "Action"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Documentaries|Docuseries"), 'listed_in'] = "Documentaries"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("Thrillers"), 'listed_in'] = "Thrillers"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("TV Shows|Reality TV"), 'listed_in'] = "TV Shows"
netflix_titles.loc[netflix_titles['listed_in'].str.contains("International Movies|Independent Movies"), 'listed_in'] = "Movies"
```

```python
netflix_titles["rating"].value_counts()
```

```python
rating_mapping = {
    'TV-MA': 'Adult',
    'TV-14': 'Teen',
    'R': 'Adult',
    'TV-PG': 'Family',
    'PG-13': 'Teen',
    'PG': 'Family',
    'NR': 'Unrated',
    'TV-G': 'Family',
    'TV-Y7': 'Children',
    'TV-Y': 'Children',
    'G': 'Family',
    'TV-Y7-FV': 'Children',
    'UR': 'Unrated',
    'NC-17': 'Adult'
}
netflix_titles['rating'] = netflix_titles['rating'].map(rating_mapping)
```

```python
netflix_titles["country"] = netflix_titles["country"].replace("United Kingdom,", "United Kingdom")
```

```python
netflix_titles = netflix_titles.drop_duplicates()
```

```python
netflix_titles.reset_index(drop = True, inplace = True)
```

```python
value_counts = netflix_titles['listed_in'].value_counts()
percentages = (value_counts / len(netflix_titles)) * 100


print("Percentages of each category in 'listed_in' column:\n", percentages)
```
